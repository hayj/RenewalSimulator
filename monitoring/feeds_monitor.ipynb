## How many news crawled per day and what are the sources?

```python
from systemtools.basics import *
from systemtools.file import *
from systemtools.printer import *
from databasetools.mongo import *
from datetime import datetime as pdtime
from datetime import datetime
from urllib.parse import urlparse
```

```python
articles = MongoCollection("renewal", "articles")
```

```python
start = datetime(2021, 9, 1)
print(start)
now = datetime.now()
print(now)
```

```python
# TODO replace last_seen by publish_date for the request filter
publish_date_field = "publish_date"
data = []
for current in articles.find({"last_seen": {'$gte': start}}, projection={"content": 0, "summary": 0, "text": 0}):
    if publish_date_field in current and current[publish_date_field] is not None:
        date = current[publish_date_field]
        if date > start and date < now:
            data.append(current)
```

```python
count = dict()
already_seen_urls = set()
netlocs = dict()
for current in data:
    date = current[publish_date_field]
    date = date.replace(hour=0, minute=0, second=0, microsecond=0)
    url = current["url"]
    if date > start and url not in already_seen_urls and date < now:
        if date not in count:
            count[date] = 0
        count[date] += 1
        netloc = urlparse(url).netloc
        if date not in netlocs:
            netlocs[date] = set()
        netlocs[date].add(netloc)
        already_seen_urls.add(url)
bp(count)
```

```python
for e in sorted(count.items(), key=lambda x: x[0]):
    print(e)
```

```python
for e in sorted(netlocs.items(), key=lambda x: x[0]):
    print("##########")
    print(e[0])
    for current in sorted(list(e[1])):
        print(current)
    print()
```
